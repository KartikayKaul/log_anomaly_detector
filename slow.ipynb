````python
import random

temps = data['templates']

def dataloader(n: int,
             templates_list: list[dict],
             normal_prob: float=0.5
            ):
    import random
    
    # init logic
    normal_templates =[t for t in templates_list if t['category'] == 'normal']
    anomaly_templates = [t for t in templates_list if t['category'] == 'anomaly']

    # loop d'etat
    while n>0:
        if random.random() < normal_prob:
            yield random.choice(normal_templates)
        else:
            yield random.choice(anomaly_templates)
        n-=1
````

````python
def lazy_timestomper(n: int, start: str, end: str, fmt: str="%Y-%m-%dT%H:%M:%SZ"):
    start = datetime.strptime(start, fmt)
    end = datetime.strptime(end, fmt)
    total_seconds = (end - start).total_seconds()

    # random gaps 
    gaps = np.random.exponential(scale=total_seconds/n, size=n)
    current_time = start
    for gap in gaps:
        current_time += timedelta(seconds=gap)
        if current_time >= end:
            break
        yield current_time.isoformat()
````

````python
from faker import Faker

class FieldGenerator:
    def __init__(self, config):
        self.config = config

    def generate(self):
        raise NotImplementedError

# faker generator
class FakerField(FieldGenerator):
    def __init__(self, config, faker_instance):
        super().__init__(config)
        self.faker = faker_instance
    
    def generate(self):
        provider = self.config['provider']
        return getattr(faker, provider)()

class StringField(FieldGenerator):
    def generate(self):
        values = self.config['values']
        return random.choice(values)

class IntCField(FieldGenerator):
    def generate(self):
        values = self.config["values"]
        return random.choice(values)

class IntDField(FieldGenerator):
    def generate(self):
        distribution = self.config.get("distribution", "uniform")

        if distribution == "normal":
            mean = self.config["mean"]
            std = self.config["std"]
            value = np.random.normal(mean, std)
            return max(0, int(round(value)))
            
        elif distribution == "uniform":
            min_val = self.config["min"]
            max_val = self.config["max"]
            return random.randint(min_val, max_val)

class FloatField(FieldGenerator):
    def generate(self):
        distribution = self.config.get("distribution", "uniform")

        if distribution == "normal":
            mean = self.config["mean"]
            std = self.config["std"]
            value = np.random.normal(mean, std)

            # avoid negative response time
            return max(0, round(value, 2))

        elif distribution == "uniform":
            min_val = self.config["min"]
            max_val = self.config["max"]
            return round(random.uniform(min_val, max_val), 2)

FIELD_TYPE_REGISTRY = {
    "faker": FakerField,
    "str": StringField,
    "int_c": IntCField,
    "int_d": IntDField,
    "float": FloatField
}

class LogFieldEngine:
    def __init__(self, field_configs):
        self.faker = Faker() # instantiate once the Faker engine
        self.generators = {}

        for field_name, config in field_configs.items():
            field_type = config["type"]
            generator_class = FIELD_TYPE_REGISTRY[field_type]

            if field_type == "faker":
                self.generators[field_name] = generator_class(config, self.faker)
            else:
                self.generators[field_name] = generator_class(config)

    def generate_field(self, field_name):
        return self.generators[field_name].generate()

    def generate_all(self):
        return {
            field: generator.generate()
            for field, generator in self.generators.items()
        }
````
